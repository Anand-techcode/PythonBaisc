# how to create a bar chart

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
import pandas as pd
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv("bar_chart_data.csv")
df
```

```python
plt.bar(x = df["Brand"],
       height = df["Cars Listings"])  # x is x ais and heinght is Y axis
plt.show()
```

default size of the garpg is 6.4 X 4.8 and as we can see that labels are overlapping so we increase the widh

```python
plt.figure(figsize = (9,6))
plt.bar(x = df["Brand"],
       height = df["Cars Listings"])  # x is x ais and heinght is Y axis
plt.show()
```

another way to remove overlapping issue by rotating the graph

```python
plt.bar(x = df["Brand"],
       height = df["Cars Listings"])  # x is x ais and heinght is Y axis
plt.xticks(rotation = 45)
plt.show()
```

with the help of seaborn we can assign different colors to the bar

```python
plt.bar(x = df["Brand"],
       height = df["Cars Listings"],
       color = ["red","green", "blue", "white","yellow","magenta","orange"])  # x is x ais and heinght is Y axis
plt.xticks(rotation = 45)
plt.title("Car listing by brands")
plt.show()
```

```python
plt.bar(x = df["Brand"],
       height = df["Cars Listings"],
       color = ["red","green", "blue", "white","yellow","magenta","orange"])  # x is x ais and heinght is Y axis
plt.xticks(rotation = 45)
plt.title("Car listing by brands", fontsize = 16, fontweight = "bold")
plt.show()
```

```python
plt.bar(x = df["Brand"],
       height = df["Cars Listings"],
       color = ["red","green", "blue", "white","yellow","magenta","orange"])  # x is x ais and heinght is Y axis
plt.xticks(rotation = 45, fontsize = 13, fontweight = "bold")
plt.yticks(fontsize = 13)
plt.title("Car listing by brands", fontsize = 16, fontweight = "bold")
plt.show()
```

not everyone is comfortable with the python code so we can export the graph as an image

```python
plt.bar(x = df["Brand"],
       height = df["Cars Listings"],
       color = ["red","green", "blue", "white","yellow","magenta","orange"])  # x is x ais and heinght is Y axis
plt.xticks(rotation = 45, fontsize = 13, fontweight = "bold")
plt.yticks(fontsize = 13)
plt.title("Car listing by brands", fontsize = 16, fontweight = "bold")
plt.show()
plt.savefig("Usedcar.png")
```

# Pie chart

```python
pie = pd.read_csv("pie_chart_data.csv")
pie
```

```python
plt.figure(figsize = (10,8))
plt.pie(pie['Number of Cars'],
       labels = pie['Engine Fuel Type'].values,  #labels takes the sequesnce of strings providing the labels for each wedge
       autopct = '%2.f%%')  #autopercent apply percentage automatically
plt.show()
```

```python
plt.figure(figsize = (10,8))
plt.pie(pie['Number of Cars'],
       labels = pie['Engine Fuel Type'].values,
       autopct = '%2.f%%',
       textprops = {'size' : 'x-large',
                   'fontweight' : 'bold'})  #labels takes the sequesnce of strings providing the labels for each wedge
plt.show()
```

```python
plt.figure(figsize = (10,8))
plt.pie(pie['Number of Cars'],
       labels = pie['Engine Fuel Type'].values,
       autopct = '%2.f%%',
       textprops = {'size' : 'x-large',
                   'fontweight' : 'bold',
                   'rotation' : '30'})  #labels takes the sequesnce of strings providing the labels for each wedge
plt.show()
```

```python
sns.set_palette('colorblind')
```

```python
plt.figure(figsize = (10,8))
plt.pie(pie['Number of Cars'],
       labels = pie['Engine Fuel Type'].values,
       autopct = '%2.f%%',
       textprops = {'size' : 'x-large',
                   'fontweight' : 'bold',
                   'rotation' : '30',
                   'color' : 'w'})  #labels takes the sequesnce of strings providing the labels for each wedge
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (10,8))
plt.pie(pie['Number of Cars'],
       labels = pie['Engine Fuel Type'].values,
       autopct = '%2.f%%',
       textprops = {'size' : 'x-large',
                   'fontweight' : 'bold',
                   'rotation' : '30',
                   'color' : 'w'})  #labels takes the sequesnce of strings providing the labels for each wedge
plt.legend()
plt.title("Pie chart" , fontsize = 18 , fontweight = 'bold')
plt.show()
```

# Line Chart

```python
line = pd.read_csv("line_chart_data.csv")
pd.options.display.max_columns = None
pd.options.display.max_rows = None
line
```

```python
line['new_date'] = pd.to_datetime(line["Date"]) #changing date column in proper date format
```

```python
line['new_date']
```

```python
plt.plot(line['new_date'], line["GSPC500"])
plt.plot(line['new_date'], line["FTSE100"])
plt.show()
```

```python
plt.figure(figsize = (20,8))
plt.plot(line['new_date'], line["GSPC500"])
plt.plot(line['new_date'], line["FTSE100"])
plt.show()
```

```python
labels = ["S&P 500", "FTSE 100"]
plt.figure(figsize = (20,8))
plt.plot(line['new_date'], line["GSPC500"])
plt.plot(line['new_date'], line["FTSE100"])
plt.title("Line chart" , fontsize = 14 , fontweight = "bold")
plt.ylabel("Returns")
plt.xlabel("Date")
plt.legend(labels = labels , fontsize = "large")
plt.show()
```

```python
new_line = line[(line.new_date >= '2008-07-01') &
               (line.new_date <= '2008-12-31')]
new_line
```

```python
labels = ["S&P 500", "FTSE 100"]
plt.figure(figsize = (20,8))
plt.plot(new_line['new_date'], new_line["GSPC500"])
plt.plot(new_line['new_date'], new_line["FTSE100"])
plt.title("Line chart" , fontsize = 14 , fontweight = "bold")
plt.ylabel("Returns")
plt.xlabel("Date")
plt.legend(labels = labels , fontsize = "large")
plt.show()
```

# stacked area chart

```python
stack = pd.read_csv("stacked_area_chart_data.csv")
stack
```

```python
plt.stackplot(stack['Year'],
             stack['Gas'],
             stack['Petrol'],
             stack['Diesel'],
             stack['Other'])
plt.show()
```

```python
plt.figure(figsize = (12,6))
plt.stackplot(stack['Year'],
             stack['Gas'],
             stack['Petrol'],
             stack['Diesel'],
             stack['Other'])
plt.show()
```

```python
plt.figure(figsize = (12,6))
plt.stackplot(stack['Year'],
             stack['Gas'],
             stack['Petrol'],
             stack['Diesel'],
             stack['Other'])
plt.xticks(stack['Year'], rotation = 45)
plt.show()
```

```python
colors = ["011638" , "#7e2987" , "ef2026"]
plt.figure(figsize = (12,6))
plt.stackplot(stack['Year'],
             stack['Gas'],
             stack['Petrol'],
             stack['Diesel'],
             stack['Other'],
             colors = colors)
plt.xticks(stack['Year'], rotation = 45)
plt.show()
```

```python
colors = ["011638" , "#7e2987" , "ef2026"]
plt.figure(figsize = (12,6))
plt.stackplot(stack['Year'],
             stack['Gas'],
             stack['Petrol'],
             stack['Diesel'],
             stack['Other'],
             colors = colors,
             edgecolor = 'none')
plt.xticks(stack['Year'], rotation = 45)
plt.show()
```

```python
colors = ["011638" , "#7e2987" , "ef2026"]
labels = ["Gas","Petrol" , "Diesel"]
plt.figure(figsize = (12,6))
plt.stackplot(stack['Year'],
             stack['Gas'],
             stack['Petrol'],
             stack['Diesel'],
             stack['Other'],
             colors = colors,
             edgecolor = 'none')
plt.xticks(stack['Year'], rotation = 45)
plt.legend(labels = labels , loc = "upper left")
plt.show()
```

```python
colors = ["011638" , "#7e2987" , "ef2026"]
labels = ["Gas","Petrol" , "Diesel"]
plt.figure(figsize = (12,6))
plt.stackplot(stack['Year'],
             stack['Gas'],
             stack['Petrol'],
             stack['Diesel'],
             stack['Other'],
             colors = colors,
             edgecolor = 'none')
plt.xticks(stack['Year'], rotation = 45)
plt.ylabel("Number of Cars" , fontsize = 16)
plt.legend(labels = labels , loc = "upper left")
plt.title("Popularity of engine fuel")
plt.show()
```
